# APPLY K-MEANS CLUSTERING METHOD

```python
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
```

*   Let's scale the data first

```python
scaler = StandardScaler()
```

```python
scaled_creditcard = scaler.fit_transform(creditcard_df)
```

*   Find the optimal number of Clusters using Elbow Method

```python
scores_1 = []
for i in range(1,15):
    kmeans = KMeans(n_clusters=i)
    kmeans.fit(scaled_creditcard)
    scores_1.append(kmeans.inertia_)
```

```python
plt.plot(scores_1, 'bx-')
plt.title('Finding the right number of clusters')
plt.xlabel('Clusters')
plt.ylabel('Scores') 
plt.show()
```

*   From this we can observe that the values reduce with huge amounts at first and starts reducing linearly after around 8th cluster. 
*   So, let's choose the number of clusters to be 8 and apply clustering again.

```python
kmeans = KMeans(8)
kmeans.fit(scaled_creditcard)
labels = kmeans.labels_
```

```python
# let's perform inverse transformation and Put the cluster centers into dataframe
cluster_centers = scaler.inverse_transform(kmeans.cluster_centers_)
cluster_centers = pd.DataFrame(data = cluster_centers, columns = [creditcard_df.columns])
cluster_centers
```

```python
# concatenate the clusters labels to our original dataframe (add labels as new column)
creditcard_df_cluster = pd.concat([creditcard_df, pd.DataFrame({'cluster':labels})], axis=1)
```

```python
# Plot the histogram of each cluster per feature
for i in creditcard_df.columns: # For each feature
    plt.figure(figsize=(35,5))
    for j in range(8): # For each cluster
        plt.subplot(1, 8, j+1)
        cluster = creditcard_df_cluster[creditcard_df_cluster['cluster'] == j]
        cluster[i].hist(bins = 20)
        plt.title(f'{i} \nCluster {j}')
plt.show()
```

SOME CLUSTERS THAT WE CAN INTERPRET:
*   First Customers cluster (Transactors): Those are customers who pay least amount of intrerest charges and careful with their money, Cluster with lowest balance (\\$104) and cash advance (\\$305), Percentage of full payment = 18%
*   Second customers cluster (revolvers) who use credit card as a loan (most lucrative sector): highest balance (\\$5000) and cash advance (~\\$5000), low purchase frequency, high cash advance frequency (0.5), high cash advance transactions (16) and low percentage of full payment (3%)
*   Third customer cluster (VIP/Prime): high credit limit \\$16K and highest percentage of full payment, target for increase credit limit and increase spending habits
*   Fourth customer cluster (low tenure): these are customers with low tenure (7 years), low balance 

# APPLY PRINCIPAL COMPONENT ANALYSIS (PCA)

```python
from sklearn.decomposition import PCA
```

```python
# Obtain the principal components 
pca = PCA(n_components=2)
principal_comp = pca.fit_transform(scaled_creditcard)
```

```python
# Create a dataframe with the two components
pca_df = pd.DataFrame(data=principal_comp, columns=['pca1','pca2'])
```

```python
# Concatenate the clusters labels to the dataframe
pca_df = pd.concat([pca_df, pd.DataFrame({'cluster':labels})], axis=1)
pca_df.head()
```

```python
# Visualize this data
plt.figure(figsize=(10,10))
sns.scatterplot(x='pca1', y='pca2', hue='cluster', data=pca_df, palette=['red','green','blue','pink','yellow','gray','purple', 'black'])
plt.show()
```

# PERFORM DIMENSIONALITY REDUCTION USING AUTOENCODERS

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import SGD
```

```python
# ENCODER
encoder = Sequential()
encoder.add(Dense(units=12, activation='relu', input_shape=(17,)))
encoder.add(Dense(units=8, activation='relu'))
```

```python
# DECODER
decoder = Sequential()
decoder.add(Dense(units=12, activation='relu', input_shape=(8,)))
decoder.add(Dense(units=17, activation='relu'))
```

```python
autoencoder = Sequential([encoder, decoder])
```

```python
autoencoder.compile(optimizer=SGD(learning_rate=0.1), loss='mse')
```

```python
autoencoder.fit(scaled_creditcard, scaled_creditcard, epochs=25)
```

```python
# Get predictions
pred = encoder.predict(scaled_creditcard)
pred.shape
```

```python
# Find Optimal Number of Clusters (with the new reduced features)
scores_2 = []
for i in range(1,10):
    kmeans = KMeans(n_clusters=i)
    kmeans.fit(pred)
    scores_2.append(kmeans.inertia_)
```

```python
plt.plot(scores_2, 'bx-')
plt.title('Finding the right number of clusters')
plt.xlabel('Clusters')
plt.ylabel('Scores') 
plt.show()
```

*   Let's select 4 clusters

```python
kmeans = KMeans(4)
kmeans.fit(pred)
labels = kmeans.labels_
```

```python
# Add new Cluster labels to the original dataframe
df_cluster_dr = pd.concat([creditcard_df, pd.DataFrame({'cluster':labels})], axis = 1)
df_cluster_dr.head()
```
